### 5) Make sure that all statistical assumptions are satisfied (perform necessary statistical tests to ensure it)

Проверим наш ряд Prep_Rate на стационарность тестом Дика-Фуллера

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries, window = 12, cutoff = 0.05):

    #Determing rolling statistics
    rolmean = timeseries.rolling(window).mean()
    rolstd = timeseries.rolling(window).std()

    #Plot rolling statistics:
    fig = plt.figure(figsize=(12, 4))
    orig = plt.plot(np.arange(len(timeseries)),timeseries, color='blue',label='Original')
    mean = plt.plot(np.arange(len(rolmean)),rolmean, color='red', label='Rolling Mean')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    std = plt.plot(np.arange(len(rolstd)),rolstd, color='black', label = 'Rolling Std')
    
    plt.show()
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries.values,autolag='AIC' )
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    pvalue = dftest[1]
    if pvalue < cutoff:
        print('p-value = %.4f. The series is likely stationary.' % pvalue)
    else:
        print('p-value = %.4f. The series is likely non-stationary.' % pvalue)
    
    print(dfoutput)
    
test_stationarity(df['Prep_Rate'])
```

Ряд нестационарный! Поэтому, например, моделью Sarimax его нельзя предсказывать, нужно взять разности 1 порядка

### 6) Propose validation scheme for the model and conduct it. Does a model seem to be stable? What are the indicators of the model’s stability?

```python
amt_for_test=12
Xtrain=df.iloc[:-amt_for_test,[2,3]]
ytrain=df['Prep_Rate'].values[:-amt_for_test]
Xtest=df.iloc[-amt_for_test:,[2,3]]
ytest=df['Prep_Rate'].values[-amt_for_test:]

from sklearn.linear_model import LinearRegression
mymodel=LinearRegression(fit_intercept=True)
mymodel.fit(Xtrain,ytrain)
ypred=mymodel.predict(Xtest)

from sklearn.metrics import mean_absolute_error as mae
#print("ytest=",ytest)
#print("ypred=",ypred)
print("mae=",mae(ytest,ypred))
```

По значению mae кажется, что линейная регрессия научилась довольно хорошо предсказывать Pred_Rate, но посмотрим на графике

```python
plt.plot(np.arange(len(ytrain)),ytrain,label='train')
plt.plot(len(ytrain)+np.arange(len(ytest)),ytest,label='test')
plt.plot(len(ytrain)+np.arange(len(ytest)),ypred,label='pred')
plt.legend()
plt.show()
```

Теперь видно, что на самом деле не очень хорошо

### 7) Create some benchmark, and test how your model performs against it

В качестве benchmark возьмем предсказание константой

```python
c=np.mean(ytrain)
ybenchmark=[c for i in range(len(Xtest))]
print("mae=",mae(ytest,ybenchmark))

plt.plot(np.arange(len(ytrain)),ytrain,label='train')
plt.plot(len(ytrain)+np.arange(len(ytest)),ytest,label='test')
plt.plot(len(ytrain)+np.arange(len(ytest)),ypred,label='pred')
plt.plot(len(ytrain)+np.arange(len(ytest)),ybenchmark,label='benchmark')
plt.legend()
plt.show()
```

Видим, что линейная регрессия справилась все-таки лучше, чем константа, это радует
