```python
df_spt=pd.read_parquet("trades_spt.parquet",engine='pyarrow')
```

```python
df_spt
```

```python
df_spt1=df_spt[df_spt['side']==1]
df_sptm1=df_spt[df_spt['side']==-1]
```

```python
df_spt_new=df_spt.copy()
```

```python
df_spt_new['b_price']=(df_spt_new['side']+1)*df_spt_new['price']/2
df_spt_new['b_price']=df_spt_new['b_price'].replace({0:np.nan})
df_spt_new.bfill(inplace=True)

df_spt_new['s_price']=(df_spt_new['side']-1)*df_spt_new['price']/(-2)
df_spt_new['s_price']=df_spt_new['s_price'].replace({0:np.nan})
df_spt_new.bfill(inplace=True)

df_spt_new['spread']=2*(df_spt_new['b_price']-df_spt_new['s_price'])/(df_spt_new['b_price']+df_spt_new['s_price'])*10000

df_spt_new
```

```python
df_spt_new['spread'].plot()
```

```python
df_spt_new['spread'].rolling(window=10000).mean().plot()
```

```python
df_spt_new['spread'].rolling(window=10000).mean().plot()
```
