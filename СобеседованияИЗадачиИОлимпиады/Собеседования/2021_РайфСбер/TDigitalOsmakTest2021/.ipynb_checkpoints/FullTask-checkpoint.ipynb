Теперь разбираемся со спредом;
Понятно, что настоящий спред мы не получим, а получим только некоторое приближение, 
потому что мы имеем только цены, по которым мы купили-продали, но настоящий бид и аск мы не знаем
Считаем, что buy_price - это цена ближайшей в будущем цены покупки,
а sell_price - это цена ближайшей в будущем цены продажи
То есть делаем backward-fill для цен покупки и для цен продажи
И спред=2*(b_price-s_price)/(b_price+s_price)*10000
Здесь в числителе стоит 2*spread, а в знаменателе стоит две справедливых цены
Мы тут на 10000 умножили, чтобы спред измерялся в базисных пунктах

```python
df_fut['b_price']=(df_fut['side']+1)*df_fut['price']/2
df_fut['b_priceBfilled']=df_fut['b_price'].replace({0:np.nan})
df_fut.bfill(inplace=True)

df_fut['s_price']=(df_fut['side']-1)*df_fut['price']/(-2)
df_fut['s_priceBfilled']=df_fut['s_price'].replace({0:np.nan})
df_fut.bfill(inplace=True)

df_fut['spread']=2*(df_fut['b_priceBfilled']-df_fut['s_priceBfilled'])/(df_fut['b_priceBfilled']+df_fut['s_priceBfilled'])*10000
#df_fut=df_fut.replace({0:np.nan})


print(df_fut['spread'][0:15])
```

```python
df_spt['b_price']=(df_spt['side']+1)*df_spt['price']/2
df_spt['b_priceBfilled']=df_spt['b_price'].replace({0:np.nan})
df_spt.bfill(inplace=True)

df_spt['s_price']=(df_spt['side']-1)*df_spt['price']/(-2)
df_spt['s_priceBfilled']=df_spt['s_price'].replace({0:np.nan})
df_spt.bfill(inplace=True)

df_spt['spread']=2*(df_spt['b_priceBfilled']-df_spt['s_priceBfilled'])/(df_spt['b_priceBfilled']+df_spt['s_priceBfilled'])*10000
df_spt=df_spt.replace({0:np.nan})

print(df_spt['spread'][0:15])
```

Нас интересует кусок с 23:59:58.041 до 23:59:59.806;
В нашей модели там много отрицательного спреда;
Но покажем, как такие заявки могли исполняться в реальном мире, причем истинный спред всегда был положительный, 
то есть аск больше бида;
А все проблемы из-за того, что мы не учитываем время появления заявок

Итак, в момент 23:59:58.041 произошла покупка за 1969.56, и аск теперь >=1969.56, например, 1969.60

Далее в 23:59:58.261 произошла продажа за 1969.44, и бид упал (потому что остальные хотят купить дешевле, чем 1969.44), 
        и пусть теперь бид стал 1969.30

Тогда как дальше в 23:59:58.791 произошла продажа за 1969.48, это же больше бида (а мы считаем, что все сделки агрессивные, а агрессивная покупка происходит по биду)?


Очень просто: в промежутке от 23:59:58.261 до 23:59:58.791 нашелся человек, готовый купить дороже: не за 1969.30, а за 1969.48, его заявку исполнили, и теперь бид опять стал 1969.30

Дальше, в промежутке от 23:59:58.791 до 23:59:58.930 нашелся человек, который готов продать дешевле:
        не за 1969.56, а за 1969.40-1969.49;

То есть в моменты  от 23:59:58.261 до 23:59:58.791 у нас истинный спред положительный, потому что аск>=1969.56, а бид<=1969.48,
но в нашей модели он отрицательный, потому  что 1969.40-1969.49 спропагировалось назад;
Но проблема в том, что 1969.40-1969.49 спропагировалось на моменты, когда этих заявок 1969.40-1969.49 вообще еще не пришло!
То есть причина появления отрицательного спреда в том, что мы не знаем время прихода заявок

Хорошо, а почему в моменты 23:59:58.930-23:59:59.246 спред отрицательный?

Потому что в это время бид по-прежнему равен 1969.30, аск равен 1969.40-1969.49, то есть истинный спред положительный, 
но спред в нашей модели отрицательный, потому что там 1969.53 спропагировалось назад, на моменты, когда этой заявки еще не было!

А к моменту 23:59:59.282, когда заявка 1969.53 уже пришла - там уже и аск>=1969.49, то есть например 1969.60, то есть опять истинный спред положительный!

То есть отрицательный спред в модели бывает, если сейчас произошла покупка (и аск пошел вверх), 
а бОльшая цена продажи спропагировалась из будущего, то есть и бид пошел вверх, значит, и цена актива пошла вверх

Или отрицательный спред может быть получен так: сейчас произошла продажа (и бид пошел вниз), 
а мЕньшая цена покупки пришла из будущего, то есть сейчас и аск идет вниз, то есть и справедливая цена актива идет вниз

То есть у нашей модели бывает отрицательный спред в те моменты, когда цена актива сильно изменяется (идет вверх или вниз), при этом истинный спред всегда положительный

#### 5. Построить стратегию на основе сигнала из пункта (4) и сравнить ее со
стратегией buy-and-hold. Стратегия может быть построена на споте, на
фьючерсе или на их комбинации (бэктест можно строить как на тиковых данных, так и на ценах построенных баров)

Выбираем какие-нибудь пороги и считаем, сколько заработает стратегия

Покупаем или продаем всегда только одну акцию

```python
df_futGrouped.iloc[1075108,7]=1969.53
df_futGrouped.iloc[1075107,7]=1969.53
df_futGrouped.iloc[1075106,7]=1969.53
df_futGrouped.tail(5)
```

```python
a=0.5 #if less than buy
b=1 #if greater than sell

df_futGrouped['here_buy']=(df_futGrouped['spreadNew']<=a).astype(int)
df_futGrouped['here_sell']=(df_futGrouped['spreadNew']>=b).astype(int)

loss=(df_futGrouped['here_buy']*df_futGrouped['b_priceBfilled']).sum()
profit=(df_futGrouped['here_sell']*df_futGrouped['s_priceBfilled']).sum()
tek_amt=df_futGrouped['here_buy'].sum()-df_futGrouped['here_sell'].sum()

last_buy_price=(df_futGrouped[df_futGrouped['side']==1]['price']).values[-1]
last_sell_price=(df_futGrouped[df_futGrouped['side']==-1]['price']).values[-1]
last_middle_price=(last_buy_price+last_sell_price)/2
pnl=profit-loss+tek_amt*last_middle_price

print('pnl=',pnl)
print('проторгованный объем=',df_futGrouped['here_buy'].sum()+df_futGrouped['here_sell'].sum())
print('маржинальность=',pnl/(df_futGrouped['here_buy'].sum()+df_futGrouped['here_sell'].sum()))
```

Посторим график pnl стратегии с течением времени (и убедимся, что в последний момент pnl такой же, как мы посчитали выше)

```python
trajectory_loss=(df_futGrouped['here_buy']*df_futGrouped['b_priceBfilled']).cumsum()
trajectory_profit=(df_futGrouped['here_sell']*df_futGrouped['s_priceBfilled']).cumsum()
trajectory_tek_amt=(df_futGrouped['here_buy']-df_futGrouped['here_sell']).cumsum()
trajectory_middle_price=(df_futGrouped['s_priceBfilled']+df_futGrouped['b_priceBfilled'])/2
trajectory_pnl=trajectory_profit-trajectory_loss+trajectory_tek_amt*trajectory_middle_price
print(trajectory_pnl.values[-1])
```

Запихиваем это в функцию, и кроме pnl возвращаем еще маржинальность=pnl/проторгованный объем

```python
def get_pnl(a,b):
    df_futGrouped['here_buy']=(df_futGrouped['spreadNew']<=a).astype(int)
    df_futGrouped['here_sell']=(df_futGrouped['spreadNew']>=b).astype(int)
    loss=(df_futGrouped['here_buy']*df_futGrouped['b_priceBfilled']).sum()
    profit=(df_futGrouped['here_sell']*df_futGrouped['s_priceBfilled']).sum()
    tek_amt=df_futGrouped['here_buy'].sum()-df_futGrouped['here_sell'].sum()
    last_buy_price=(df_futGrouped[df_futGrouped['side']==1]['price']).values[-1]
    last_sell_price=(df_futGrouped[df_futGrouped['side']==-1]['price']).values[-1]
    last_middle_price=(last_buy_price+last_sell_price)/2
    pnl=profit-loss+tek_amt*last_middle_price
    return pnl/(df_futGrouped['here_buy'].sum()+df_futGrouped['here_sell'].sum())
```

Для разных a и b ищем маржинальность:

```python
otv=pd.DataFrame(columns=['a','b','marg'])
```

```python
for a in [0.1,0.3,0.5,0.7,1,1.5,2]:
    for b in [0.1,0.3,0.5,0.7,1,1.5,2]:
        if (a <b):
            otv=otv.append({'a':a,'b':b,'marg':get_pnl(a,b)},ignore_index=True)
            print(a,b,get_pnl(a,b))
```

Найдем параметры, дающие максимальную маржинальность:

```python
def read(myname):
    df=pd.read_parquet(myname, engine='pyarrow')
    df['timestamp']=pd.to_datetime(df['exch_ts'], unit='ns')
    df.set_index('timestamp', inplace=True)
    return df
```

Пересэмплируем все по 1 секунде, потому что мы для спота будем смотреть значение фьючерса в этот момент, а моменты у спота и у фьючерса не обязаны совпадать, а так мы их принудительно заставим совпадать

```python
def get_price(df,myside):
    price = df[df['side'] == myside]['price'].resample('1S').mean()
    return price.interpolate()
```

```python
def get_stats(df,need_spread):
    dfNew=pd.DataFrame(columns=['b_price','s_price'])
    dfNew['b_price']=get_price(df,1)
    dfNew['s_price']=get_price(df,-1)
    dfNew['middle_price']=(dfNew['b_price']+dfNew['s_price'])/2
    if (need_spread):
        dfNew['spread']=2*(dfNew['b_price']-dfNew['s_price'])/(dfNew['b_price']+dfNew['s_price'])*10000
    return dfNew
```

```python
df_spt=read("trades_spt.parquet")
df_spt=get_stats(df_spt,1)
df_spt.head(5)
```

```python
df_fut=read("trades_fut.parquet")
df_fut=get_stats(df_fut,0)
df_fut.bfill(inplace=True)
df_fut=df_fut.iloc[1:,:]
df_fut.head(5)
```

```python
a=0
df_spt['here_SpreadGreater']=(df_spt['spread']>=a).astype(int)
df_spt['here_FutGreater']=(df_fut['middle_price']>df_spt['middle_price']).astype(int)
df_spt['here_buy']=df_spt['here_SpreadGreater']*df_spt['here_FutGreater']
df_spt['here_sell']=(df_spt['here_SpreadGreater'])*(1-df_spt['here_FutGreater'])
df_spt.head(5)
```

```python
trajectory_loss=(df_spt['here_buy']*df_spt['b_price']).cumsum()
trajectory_profit=(df_spt['here_sell']*df_spt['s_price']).cumsum()
trajectory_tek_amt=(df_spt['here_buy']-df_spt['here_sell']).cumsum()
trajectory_middle_price=(df_spt['s_price']+df_spt['b_price'])/2
trajectory_pnl=trajectory_profit-trajectory_loss+trajectory_tek_amt*trajectory_middle_price
print(trajectory_tek_amt.values[-1])
trajectory_pnl.plot()
```

```python
def get_pnl(df_spt,df_fut,a):
    df_spt['here_SpreadGreater']=(df_spt['spread']>=a).astype(int)
    df_spt['here_FutGreater']=(df_fut['middle_price']>df_spt['middle_price']).astype(int)
    df_spt['here_buy']=df_spt['here_SpreadGreater']*df_spt['here_FutGreater']
    df_spt['here_sell']=(df_spt['here_SpreadGreater'])*(1-df_spt['here_FutGreater'])
    
    trajectory_loss=(df_spt['here_buy']*df_spt['b_price']).cumsum()
    trajectory_profit=(df_spt['here_sell']*df_spt['s_price']).cumsum()
    trajectory_tek_amt=(df_spt['here_buy']-df_spt['here_sell']).cumsum()
    trajectory_middle_price=(df_spt['s_price']+df_spt['b_price'])/2
    trajectory_pnl=trajectory_profit-trajectory_loss+trajectory_tek_amt*trajectory_middle_price
    return trajectory_pnl.values[-1]
```

```python
for a in np.linspace(0,10,20,endpoint=False):
    print(a,get_pnl(df_spt,df_fut,a))
```

Видим, что новая стратегия хотя бы в минус не уходит!

Теперь еще можно угадать время экспирации у фьючерса, то есть на какое время вперед он; для этого цены спота надо сдвигать, и при каком сдвиге ковариация будет максимальна - тот сдвиг и есть время экспирации

```python
def get_covar(x):
    sdvig_fut=df_fut['middle_price'].diff()
    sdvig_spt=df_spt['middle_price'].diff()
    return (sdvig_fut[x:]*sdvig_spt[-x:]).sum()
```

```python
%time
mas = np.vectorize(get_covar)(np.arange(49*len(df_fut)//100, 51*len(df_fut)//100))
```

```python
plt.plot(np.arange(49*len(df_fut)//100, 51*len(df_fut)//100),mas)
```
