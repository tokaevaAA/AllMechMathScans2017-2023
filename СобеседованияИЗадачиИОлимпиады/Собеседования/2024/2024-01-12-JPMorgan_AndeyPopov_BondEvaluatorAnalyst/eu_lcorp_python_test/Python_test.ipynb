Part I: IRR CALCULATIONS
    
CF_sample.csv file contains Cash Flows data of underlying loans (A,B and C) packed into a collateralized security. 
Create a python function to calculate Internal Rate of Return of Each Loan (XIRR). What XIRR value did you get for each loan? Show your output in pandas dataframe

```python
import numpy as np
import pandas as pd
import datetime
from scipy import optimize
from scipy import interpolate
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")

df=pd.read_csv('CF_sample.csv')

def get_pv_from_cashflows(df,irr):
        disc_factors=np.array([(1+irr)**(-t) for t in df['t']])
        return sum(df['Cash_Flows']*disc_factors)
    
def get_irr(df):
    asofdate=datetime.datetime.strptime(df['date'].values[0],"%m/%d/%Y").date()
    df['t']=list(map(lambda x: (datetime.datetime.strptime(x,"%m/%d/%Y").date()-asofdate).days/365,df['date'].values))
    opt_func=lambda x: get_pv_from_cashflows(df,irr=x)
    irr=optimize.root_scalar(opt_func,bracket=[-0.5,0.5]).root
    return irr

for loan_name in ['A','B','C']:
    df_tek=df[df['security']==loan_name]
    irr_tek=get_irr(df_tek)
    print("loan_name=",loan_name,'irr=',irr_tek*100,'%')
```

4.1. Define Class object 'Portfolio' and its initial parameters (cash balance, number of long holdings, number of short holdings, vector of portfolio holdings)  using '__init__' function:

```python
df=pd.read_csv('bonds_trading_data.csv')
df.head()
```

```python
class Portfolio:
    def __init__(self,cash_balance,nb_long_holdings, nb_short_holdings, bonds_list,portfolio_holdings):
        self.cash_balance=cash_balance
        self.nb_long_holdings=nb_long_holdings
        self.nb_short_holdings=nb_short_holdings
        self.bonds_list=bonds_list
        self.portfolio_holdings=portfolio_holdings
        for el in bonds_list:
            if el not in portfolio_holdings:
                portfolio_holdings[el]=0
        cnt_long=0
        cnt_short=0
        for k,v in portfolio_holdings.items():
            if v > 0:
                cnt_long+=v
            else:
                cnt_short+=v
        if cnt_long!=nb_long_holdings:
            raise LogicError('cnt_long!=nb_long_holdings')
        if cnt_short!=nb_short_holdings:
            raise LogicError('cnt_short!=nb_short_holdings')
    def get_signals(self,data):
        data['spread,bp']=(data['yield']-data['curve interpolated yield'])*100
        data['buy']=list(map(lambda x: int(x), data['spread,bp'] >= 10))
        data['close']=list(map(lambda x: int(x), data['spread,bp'] < 5))
        return data
    def buy_new_assets(self,data):
        what_to_buy=data[data['buy']==1]
        for i in range(len(what_to_buy)):
            isin=what_to_buy['ISIN'].values[i]
            price=what_to_buy['price'].values[i]
            if self.nb_long_holdings < 3 and self.cash_balance >=price*1 and self.portfolio_holdings[isin]*price <=35000:
                amt=1
                self.cash_balance-=price*amt
                self.nb_long_holdings+=amt
                self.portfolio_holdings[isin]+=amt
    def close_positions(self,data):
        print("cb before close",self.cash_balance)
        for k,v in self.portfolio_holdings.items():
            if v > 0 and data[data['ISIN']==k]['close'].values[0]==1:
                print('Close',k)
                amt=1
                price=data[data['ISIN']==k]['price'].values[0]
                #print("price=",price)
                self.cash_balance+=price*amt
                self.nb_long_holdings-=amt
                self.portfolio_holdings[k]-=amt
        print("cb after close",self.cash_balance)
    def get_pv(self,month,data):
        data=data[data['Month']==month]
        pv=self.cash_balance
        for k,v in self.portfolio_holdings.items():
            pv+=v*data[data['ISIN']==k]['price'].values[0]
        return pv
    def run_backtest(self,df):
        mas_pv=[]
        df=self.get_signals(df)
        for m in range(7,13):
            print("Start of month:",m)
            data=df[df['Month']==m]
            self.close_positions(data)
            #print(myportfolio.portfolio_holdings)
            self.buy_new_assets(data)
            #print(myportfolio.portfolio_holdings)
            pv=self.get_pv(m,data)
            print("pv=",pv)
            print('\n')
            mas_pv.append(pv)
        return mas_pv
            
```
