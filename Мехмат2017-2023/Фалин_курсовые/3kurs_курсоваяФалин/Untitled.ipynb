```python
import numpy as np
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(8, 4))

for m in [1.1,1.5,1.9,2.2]:
    x=np.linspace(0, 10,100,endpoint=True)
    y = x**m
    ax1.plot(x,y,label='m='+str(m))  
ax1.set_title('t^m выпукла вниз',fontsize=16) 
ax1.set_xlabel('x')
ax1.legend()

for m in [1.1,1.5,1.9,2.2]:
    x=np.linspace(0, 10,100,endpoint=True)
    y = 1-(1-x)**m
    ax2.plot(x,y,label='m='+str(m))  
ax2.set_title('1- (1-t)^m выпукла вверх',fontsize=16) 
ax2.set_xlabel('x')
ax2.legend()
plt.savefig('pictNEW.pdf', format='pdf', dpi=100)
```
