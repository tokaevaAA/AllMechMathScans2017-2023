```python
import numpy as np
import matplotlib.pyplot as plt
N=5000
NITER=5000
b=0.5
drob=np.zeros(NITER)
chisl=np.zeros(NITER)
znam=np.zeros(NITER)
for tek_iter in range(NITER):
    eps=np.random.randn(N)
    u=np.zeros(N)
    ksi=np.zeros(N)
    eta=np.zeros(N)
    for t in range(1,N):
        u[t]=u[t-1]*b + eps[t]
    for t in range(1,N):
        ksi[t]=(eps[t])*(u[t-1])
    for t in range(1,N):
        eta[t]=(u[t])*(u[t])
    drob[tek_iter]=ksi.sum()/eta.sum()
    chisl[tek_iter]=ksi.sum()
    znam[tek_iter]=eta.sum()

print(drob.mean())
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.hist(drob,density=True,bins=20)
plt.subplot(1,3,2)
plt.hist(chisl/N,density=True,bins=20)
plt.subplot(1,3,3)
plt.hist(znam/N,density=True,bins=20)
plt.show()
```
