```python
import numpy as np
import matplotlib.pyplot as plt
import math

def give_drob_n(n):
    eps=np.random.randn(n)
    u=np.zeros(n)
    ksi=np.zeros(n)
    eta=np.zeros(n)
    for t in range(1,n):
        u[t]=u[t-1]*b + eps[t]
    for t in range(1,n):
        ksi[t]=(eps[t])*(u[t-1])
    for t in range(1,n):
        eta[t]=(u[t])*(u[t])
    return ksi.sum()/eta.sum()

N=100
b=0.5
otv=np.zeros(N)
eps=0.01
for n in range(0,N):
    s=0
    for i in range(1000):
        tek=give_drob_n(100+5*n)
        if (math.fabs(tek) < eps):
            s=s+1
    otv[n]=s/1000
    
print(otv[0:10])
    
        
    
```

```python
print(otv[99])
plt.plot(np.arange(N),otv)
plt.show()
```

```python
def give_eta_n(n):
    eps=np.random.randn(n)
    u=np.zeros(n)
    eta=np.zeros(n)
    for t in range(1,n):
        u[t]=u[t-1]*b + eps[t]
    for t in range(1,n):
        eta[t]=(u[t])*(u[t])
    return eta.mean()

print(give_eta_n(10000000))


    
        
    
```

```python
N=100
b=0.5
otv=np.zeros(N)
eps=0.01
for n in range(0,N):
    s=0
    for i in range(1000):
        tek=give_eta_n(1000+5*n)
        if (math.fabs(tek-1.33) < eps):
            s=s+1
    otv[n]=s/1000
    
print(otv[0:10])
```

```python
print(otv[99])
plt.plot(np.arange(N),otv)
plt.show()
```
