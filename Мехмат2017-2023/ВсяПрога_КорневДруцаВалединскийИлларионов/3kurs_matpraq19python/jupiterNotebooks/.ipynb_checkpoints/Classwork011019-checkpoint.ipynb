```python
import matplotlib.pyplot as plt
X=np.linspace(-np.pi, np.pi,256,endpoint=True)
C,S=np.cos(X),np.sin(X)

plt.plot(X,C,label='Cos')
plt.plot(X,S,color='r',ls='--',lw=5, label='Sin')
plt.title('Sin and Cos',fontsize=16)
plt.xlabel('X')
plt.legend()
plt.show()
```
