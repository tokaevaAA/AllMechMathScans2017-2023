```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def u0(x):
    return np.sin(x)

mas_koef=[]
mas_x=[x0 for x0 in np.linspace(-1,1,4)]
for x0 in mas_x:
    mas_koef.append((-np.sin(u0(x0)), x0 ))
print(mas_koef)

plt.figure(figsize=(10,5))
for k,b in mas_koef:
    mas_t =[t for t in np.linspace(-2,2,10)]
    mas_x=[k*t + b for t in mas_t]
    plt.plot(mas_x, mas_t)
    
plt.axvline()
plt.axhline()
plt.show()
```

```python
def u0(x):
    return x

mas_koef=[]
mas_x=[x0 for x0 in np.linspace(-1,1,4)]
for x0 in mas_x:
    mas_koef.append((4*u0(x0)**3, x0 ))
print(mas_koef)

plt.figure(figsize=(10,5))
for k,b in mas_koef:
    mas_t =[t for t in np.linspace(-2,2,10)]
    mas_x=[k*t + b for t in mas_t]
    plt.plot(mas_x, mas_t)
    
plt.axvline()
plt.axhline()
plt.show()
```
