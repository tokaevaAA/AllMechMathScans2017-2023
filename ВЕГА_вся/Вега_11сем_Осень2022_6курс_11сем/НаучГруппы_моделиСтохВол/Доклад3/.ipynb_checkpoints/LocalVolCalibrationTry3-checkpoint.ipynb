Consider the local volatility model:
$$
dS_t = S_t r\,dt + S_t \sigma(t, S_t)\,dW_t,\\
$$

Recall the Dupire's formula, that maps call market prices to: 

$$
\sigma(t, s) = \sqrt{\frac{2C^{}_{T}(t,s)+rsC^{}_{K}(t,s) }{s^{2}C^{}_{KK}(t,s)}}
$$

The formula above includes derivatives of call price w.r.t. strike and time.
Since we know the market call prices only at specific points, interpolation is needed to recover the local volatility function.
We will use bivariate spline to interpolate call prices surface.
Spline of order $k$ has continuous derivatives of orders up to $k-1$.
To obtain continuous second order derivative we need to use cubic bivariate spline for interpolation.

# Now lets fit local_vol_model to observed prices

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sps

from ipywidgets import interact
from ipywidgets import widgets
from tqdm.auto import tqdm

from dataclasses import dataclass
from typing import Union, Callable, Optional
from copy import deepcopy
from scipy.optimize import root_scalar, brentq
from dataclasses import dataclass

from scipy.interpolate import RectBivariateSpline
#import StochVolLib as SVLib
#import blackscholes

import warnings
from  scipy.stats import norm
warnings.filterwarnings("ignore")
from scipy.interpolate import CubicSpline
from tqdm import tqdm

#about super().__init__() https://habr.com/ru/company/skillfactory/blog/683744/

@dataclass
class StockOption: #общий класс для опционов
    strike_price: Union[float, np.ndarray]
    expiration_time: Union[float, np.ndarray]  # in years
    is_call: bool

@dataclass
class CallStockOption(StockOption): #класс для коллов
    def __init__(self, strike_price, expiration_time): 
        super().__init__(strike_price, expiration_time, True) #вызов метода __init__ из базового класса StockOption, т.е. вызываем колл     
        

@dataclass
class PutStockOption(StockOption):
    def __init__(self, strike_price, expiration_time):
        super().__init__(strike_price, expiration_time, False)
        
@dataclass
class MarketState:
    stock_price: Union[float, np.ndarray]
    interest_rate: Union[float, np.ndarray]  # r, assume constant
        
        
def make_local_vol_func(calls: CallStockOption,
                        call_prices: np.ndarray,
                        interest_rate: float = 0) -> Callable[[float, float], float]:
    r=interest_rate
    T=calls.expiration_time #T len=15
    K=calls.strike_price #K len=16
    C=call_prices #15*16
    C_T=np.gradient(C,T,axis=0) #T-for x_i,C-for fi
    C_K=np.gradient(C,K,axis=1) #K-for x_i,C-for fi
    C_KK=np.gradient(C_K,K,axis=1)
    sigma=(2*C_T+r*K*C_K)/(K*K*C_KK+1e-10)
    sigma[sigma<0]=0
    sigma=np.sqrt(sigma)
    interpolation=RectBivariateSpline(T,K,sigma)

    def local_vol_func(t, s):
        return interpolation(t,s,grid=False)
    
    return local_vol_func





```

```python
def dt(option: mlvf.StockOption):
    return np.maximum(option.expiration_time, np.finfo(np.float64).eps)

def d1(option: mlvf.StockOption, state: mlvf.MarketState, vola: float):
    return 1 / (vola * np.sqrt(dt(option)))\
                * (np.log(state.stock_price / option.strike_price)
                   + (state.interest_rate + vola ** 2) * dt(option))

def d2(option: mlvf.StockOption, state: mlvf.MarketState, vola: float):
    return d1(option, state, vola) - vola * np.sqrt(dt(option))
    
def price(option: mlvf.StockOption, state: mlvf.MarketState, vola: float):
    discount_factor = np.exp(-state.interest_rate * (dt(option)))
    if option.is_call:
        return sps.norm.cdf(d1(option, state, vola)) * state.stock_price\
            - sps.norm.cdf(d2(option, state, vola)) * option.strike_price * discount_factor
    
    return sps.norm.cdf(-d2(option, state, vola)) * option.strike_price * discount_factor\
        - sps.norm.cdf(-d1(option, state, vola)) * state.stock_price

def vega(option: mlvf.StockOption, state: mlvf.MarketState, vola: float):
    return state.stock_price * sps.norm.pdf(d1(option, state, vola)) * np.sqrt(dt(option))

def calc_iv(option: mlvf.CallStockOption, state: mlvf.MarketState, option_price: float):
    
    RTOL = 1e-16
    ATOL = 1e-16
    VOL_MIN = 0.0
    VOL_MAX = 1.0
    MAX_ITER = 100
    
    vola = 0.2
    p = price(option, state, vola)

    lb = VOL_MIN
    rb = VOL_MAX
    
    for i in range(MAX_ITER):
        
        resid = price(option, state, vola) - option_price
        if abs(resid) < ATOL and abs(resid) / (option_price + 1e-4) < RTOL:
            break
        
        newton_ok = False
        
        # try Newton
        v = vega(option, state, vola)
        if abs(v) > 1e-4:
            new_vola = vola - (p - option_price) / v
            if lb < new_vola < rb:
                newton_ok = True
        
        # if something went wrong, bisect
        if not newton_ok:
            new_vola = (lb + rb) / 2

        new_p = price(option, state, new_vola)
        if new_p > option_price:
            rb = new_vola
        else:
            lb = new_vola
        
        vola = new_vola
        p = new_p
        
    return vola
```

```python
def LV_func_by_IV(calls: CallStockOption,
                  T: np.ndarray,
                  K: np.ndarray,
                  call_prices: np.ndarray,
                  market: MarketState,
                  IV: np.ndarray) -> np.ndarray:
    #-> Callable[[float, float], float]:
    T_market=calls.expiration_time 
    K_market=calls.strike_price 
    #market = MarketState(stock_price=1.0, interest_rate=0.0)
    r = market.interest_rate
    s_0 = market.stock_price
    #IV = call_iv(market.stock_price, market.interest_rate, call_prices, T_market, K_market)
    # найдем y_i(K) (страница 4 в Historical Backstesting of Local Vol Model using AUD/USD ...)
    # (T, K)
    Theta = np.zeros((len(T), len(K)))
    dTheta_dT = np.zeros((len(T), len(K)))
    dTheta_dK = np.zeros((len(T), len(K)))
    d2Theta_dK2 = np.zeros((len(T), len(K)))
    tmp = np.zeros((len(T_market), len(K)))
    tmp1 = np.zeros((len(T_market), len(K)))
    tmp2 = np.zeros((len(T_market), len(K)))
    
    #print(len(T_market))
    for i in tqdm(range(len(T_market))):
        y_i = CubicSpline(K_market, IV[i,:], bc_type = 'natural')
        der_y_i = y_i.derivative(nu=1)
        der2_y_i = y_i.derivative(nu=2)
        #print(y_i(K))
        tmp[i, :] = y_i(K)
        
        tmp1[i, :] = der_y_i(K)
        tmp2[i, :] = der2_y_i(K)

    for j in tqdm(range(len(K))):
        func_dTheta_dK = CubicSpline(T_market, tmp1[:, j], bc_type = 'natural')
        dTheta_dK[:, j] = func_dTheta_dK(T) 
        func_d2Theta_dK2 = CubicSpline(T_market, tmp2[:, j], bc_type = 'natural')
        d2Theta_dK2[:, j] = func_d2Theta_dK2(T)
        func_theta = CubicSpline(T_market, tmp[:, j], bc_type = 'natural')
        Theta[:, j] = func_theta(T)
        func_dTheta_dT = func_theta.derivative(nu=1)
        dTheta_dT[:, j] = func_dTheta_dT(T)
    
    sigma2_LV = np.zeros((len(T), len(K)))
    #print(T)
    for i in tqdm(range(len(T))):
        for j in range(len(K)):
            sigma2_LV[i,j] = Theta[i,j] * Theta[i,j] + 2*T[i]*Theta[i,j]*(dTheta_dT[i,j] + r*K[j]*dTheta_dK[i,j])
            #sigma2_LV[i,j] = 2*T[i]*Theta[i,j]*(dTheta_dT[i,j] + r*K[j]*dTheta_dK[i,j])
            d_1 = (np.log(s_0/K[j]) + (r + 0.5*Theta[i,j])*T[i])
            d_1 /= Theta[i,j]*np.sqrt(T[i])
            sigma2_LV[i,j] /= np.square(1 + d_1*K[j]*np.sqrt(T[i])*dTheta_dK[i,j]) + np.square(K[j])*Theta[i,j]*T[i]*(d2Theta_dK2[i,j] - d_1*np.sqrt(T[i])*np.square(dTheta_dK[i,j]))
            sigma2_LV[i,j] = np.max(sigma2_LV[i,j], 0)
    
    sigma_LV = np.sqrt(sigma2_LV)
    
    return sigma_LV

```

```python
IV = np.empty_like(observed_call_prices) #market implied vol
for i in range(len(observed_calls.expiration_time)):
    for j in range(len(observed_calls.strike_price)):
        IV[i, j] = calc_iv(option=mlvf.CallStockOption(observed_calls.strike_price[j], observed_calls.expiration_time[i]), 
                           state=observed_market_state, 
                           option_price=observed_call_prices[i, j])
```

```python
T_max = np.max(observed_calls.expiration_time)
gamma = 1 # from paper

ind = np.argmin(np.abs(observed_calls.strike_price - 1.))
ATM_iv = IV[:, ind] # BS implied vol of ATM market options
D = gamma * np.mean(ATM_iv) * np.sqrt(T_max)

N = int(200*T_max + 200)
M = N
dx = 2*D / M
dt = T_max / N 
times = np.linspace(0.3, T_max, N+1)
S_0 = observed_market_state.stock_price
strikes = S_0 * np.exp(np.linspace(-D, D, M+1))

K, T = np.meshgrid(strikes, times)

loc_vol_func = LV_func_by_IV(observed_calls, times, strikes, observed_call_prices, observed_market_state, IV)
```

```python
def make_local_vol_func(calls: CallStockOption,
                        call_prices: np.ndarray,
                        interest_rate: float = 0) -> Callable[[float, float], float]:
    """Returns local volatility function given market call prices
    
    Important note: the arguments of the spline call can be interpreted
    two different ways base on boolean parameter ``grid``. If set to True,
    ``x`` and ``y`` (if are 1-dimenstional arrays) are broadcasted to
    meshgrid and the spline will be evaluated in grid points. In this
    case the values in arrays must be sorted. But for compatibility with
    other code, parameter ``grid`` shoud be set to False. In this case the
    spline will be evaluated in points (x[i], y[i]), which is what we need.
    
    Args:
        calls (CallStockOption): an object of type CallStockOption.
            The attributes are the points at which market call prices are observed.
        call_prices (np.ndarray): matrix of call prices evaluated at given points.
            The dimensions are (time, strikes).
        interest_rate (float): the risk-free rate.
        
    Returns:
        function that evaluates local volatility (not variance!) at given point
    """
    
    # It is recommended, first, to caculate numerical derivatives (use np.gradient),
    # evaluate the LV function at given points and only then interpolate
    # YOU CODE HERE
    
    r = interest_rate
    
    C_T_num = np.gradient(call_prices, calls.expiration_time, axis = 0)
    C_K_num = np.gradient(call_prices, calls.strike_price,  axis = 1)
    C_KK_num = np.gradient(C_K_num, calls.strike_price, axis = 1)
    
    local_vol_num = np.zeros((len(calls.expiration_time), len(calls.strike_price)))
    
    for i in range (len(calls.expiration_time)):
        for j in range (len(calls.strike_price)):
            s = calls.strike_price[j]
            local_vol_num[i][j] = (2*C_T_num[i][j] + 2*r*s*C_K_num[i][j]) / (s*s*C_KK_num[i][j] + 1e-10)
    
    local_vol_num = np.maximum(local_vol_num, 0)
    local_vol_num = np.sqrt(local_vol_num)
             

    spline = RectBivariateSpline(calls.expiration_time, calls.strike_price, local_vol_num)
    
    def local_vol_func(t, s):
        
        # YOU CODE HERE
        # in this funciton you can use all arguments of enclosing funciton
        # use `grid=False` in call to the spline    
        
        return spline.__call__(t, s, grid = False)          
        
        raise NotImplementedError
    
    return local_vol_func
```

```python
def a_current(r, i, dx, dt, loc_vol):  #t - фиксированный момент времени = t_i
    a = 0.5 / dx / dx * loc_vol[i, :] + 0.5 * r + 1 / dt  
    return a

def b_current(r, i, dx, dt, loc_vol, nu):
    b = 0.25 / dx / dx * np.square(loc_vol[i, :]) + 0.25 / dx * nu[i, :]
    return b

def c_current(r, i, dx, dt, loc_vol, nu):
    c = 0.25 / dx / dx * np.square(loc_vol[i, :])  - 0.25 / dx * nu[i, :]
    return c

def d_current(r, i, dx, dt, loc_vol):
    d = -0.5 / dx / dx * np.square(loc_vol[i, :]) - 0.5 * r + 1. / dt
    return d
    
def right_side(V_prev, r, i, dx, dt, loc_vol, nu):   # вектор с индексами по j  1, ..., M-1
    
    d = d_current(r, i, dx, dt, loc_vol)
    b = b_current(r, i, dx, dt, loc_vol, nu)
    c = c_current(r, i, dx, dt, loc_vol, nu)
    
    #print('d.shape = ', d.shape)
    #print(V_prev.shape)
    f = d[1:M] * V_prev[1:M] + b[1:M] * V_prev[2:M+1] + c[1:M] * V_prev[0:M-1]
    
    return f

def progonka(M, f, c, d, a):
    alpha = np.zeros(M+1)
    beta = np.zeros(M+1)
    alpha[1] = 1.
    beta[1] = f[0] 
    
    for k in range(1, M):
        alpha[k+1] = d[k] / (c[k] - alpha[k] * a[k])
        beta[k+1] = (f[k] + a[k] * beta[k]) / (c[k] - alpha[k] * a[k])
        
    y = np.zeros(M+1)
    y[M] = (f[M] + a[M] * beta[M]) / (1. - alpha[M])
    
    for k in range(N-1, -1, -1):
        y[k] = alpha[k+1] * y[k+1] + beta[k+1]
        
    return y
```

```python
def V_calculation(M, N, S, T, loc_vol, K, D,
                  market: MarketState):
    """
    методом прогонки вычиcляет решение сеточной задачи

    """
    #(t - первая ось, S - вторая)
    # t = T
    # S - массив, в котором лежат точки разбиения по S, т.е. s_j, j = 0, ..., M
    dx = 2*D / M
    dt = T / N
    #print(1/dt)
    r = market.interest_rate
    #print('r = ', r)
    V = np.zeros((N+1, M+1))
    V[N, :] = np.max(S - K, 0) # K - фиксированный страйк
    
    nu = - 0.5 * np.square(loc_vol) + r
    #теперь с N-ого слоя должны перейти на N-1
    # вычисляем правую часть по имеющимся значениям V и sigma
    for i in range(N, 0, -1):
        f = np.zeros(M+1)  # правая сторона из разностной схемы, ьерем её и применяем метод прогонки
        f[1:M] = right_side(V[i, :], r, i, dx, dt, loc_vol, nu)
        f[0] = 0.0            # потому что опцион call
        f[M] = S[M] - S[M-1]
        
        a = a_current(r, i-1, dx, dt, loc_vol)
        b = b_current(r, i-1, dx, dt, loc_vol, nu)
        c = c_current(r, i-1, dx, dt, loc_vol, nu)
        V[i-1, :] = progonka(M, f, a, c, b)
        
    return V
```

```python
K = 1.2
V = V_calculation(M, N, strikes, T_max, loc_vol_func, K, D, observed_market_state)
```
