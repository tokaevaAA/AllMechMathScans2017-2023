```python
import vol.vol
from vol.vol import heston
import numpy as np
import matplotlib.pyplot as plt
```

```python
myheston=heston.Heston(s=100,v=1,kappa=1, theta=0.4,sigma=0.4,rho=0.1,r=0.05)
Ts=[0.5,0.75,1,1.5,2,2.5,3]
Ks=[80,85,90,95,100,105,110,115,120]
Cs1=myheston.call_price(np.array(Ts).reshape(1,-1),np.array(Ks[0]).reshape(-1,1))
ivs1=myheston.iv(np.array(Ts).reshape(1,-1),np.array(Ks[0]).reshape(-1,1))
Cs2=myheston.call_price(np.array(Ts[0]).reshape(1,-1),np.array(Ks).reshape(-1,1))
ivs2=myheston.iv(np.array(Ts[0]).reshape(1,-1),np.array(Ks).reshape(-1,1))

fig,ax=plt.subplots(2,2)
fig.set_figheight(10)
fig.set_figwidth(10)
ax[0][0].plot(Ts,Cs1.flatten())
ax[0][0].set_xlabel('T in year_frac')
ax[0][0].set_ylabel('Call price')
ax[0][0].set_title('C(T)')

ax[1][0].plot(Ts,ivs1.flatten())
ax[1][0].set_xlabel('T in year_frac')
ax[1][0].set_ylabel('iv')
ax[1][0].set_title('iv(T)')

ax[0][1].plot(Ks,Cs2.flatten())
ax[0][1].set_xlabel('K in rub')
ax[0][1].set_ylabel('Call price')
ax[0][1].set_title('C(K)')

ax[1][1].plot(Ks,ivs2.flatten())
ax[1][1].set_xlabel('K in rub')
ax[1][1].set_ylabel('iv')
ax[1][1].set_title('iv(K)')
plt.show()
```
